```python
import numpy as np
import pandas as pd
import tensorflow as tf
import sklearn
from tqdm import tqdm
from transformers import BertTokenizer, TFBertForSequenceClassification, InputExample, InputFeatures
from sklearn.model_selection import train_test_split
```

```python
# Chargement du modèle BERT pré-entraîné et du tokenizer
model_Bert = TFBertForSequenceClassification.from_pretrained("bert-base-uncased")
tokenizer = BertTokenizer.from_pretrained("bert-base-uncased")
model_Bert.summary()
```

```python
# Transformation des étiquettes "positive" et "negative" en valeurs numériques (1 et 0)
def cat2num(value):
    if value == 'positive':
        return 1
    else:
        return 0

dataImdb['sentiment'] = dataImdb['sentiment'].apply(cat2num)
```

```python
# Séparation des données en ensembles d'entraînement (70%) et de test + validation (30%)
train, test_val = train_test_split(dataImdb, test_size=0.3, random_state=42)

# Séparation des données de test_val en ensembles de test (50%) et de validation (50%)
test, validation = train_test_split(test_val, test_size=0.5, random_state=42)
```

```python
# Enregistrement des données dans des fichiers CSV
train.to_csv("/content/drive/MyDrive/MemoireM2/Notebook/Data_Imdb/train.csv", index=False)
test.to_csv("/content/drive/MyDrive/MemoireM2/Notebook/Data_Imdb/test.csv", index=False)
validation.to_csv("/content/drive/MyDrive/MemoireM2/Notebook/Data_Imdb/validation.csv", index=False)
```

```python
# Fonction pour convertir les exemples de données en objets InputExample
def convert2inputexamples(data_input, review, sentiment):
    data_examples = data_input.apply(lambda x: InputExample(guid=None, text_a=x[review], label=x[sentiment]), axis=1)
    return data_examples

train_examples= convert2inputexamples(train, 'review', 'sentiment')
validation_examples= convert2inputexamples(validation, 'review', 'sentiment')
test_examples= convert2inputexamples(test, 'review', 'sentiment')

```

```python
# Conversion des exemples en objets TF pour l'entraînement
def convertexamples2tf(examples, tokenizer, max_length=128):
    features = []

    for i in tqdm(examples):
        input_dict = tokenizer.encode_plus(
            i.text_a,
            add_special_tokens=True,
            max_length=max_length,
            return_token_type_ids=True,
            return_attention_mask=True,
            pad_to_max_length=True,
            truncation=True
        )

        input_ids, token_type_ids, attention_mask = (input_dict["input_ids"], input_dict["token_type_ids"], input_dict['attention_mask'])
        features.append(InputFeatures(input_ids=input_ids, attention_mask=attention_mask, token_type_ids=token_type_ids, label=i.label))

    def generate():
        for f in features:
            yield (
                {
                    "input_ids": f.input_ids,
                    "attention_mask": f.attention_mask,
                    "token_type_ids": f.token_type_ids,
                },
                f.label,
            )

    return tf.data.Dataset.from_generator(
        generate,
        ({"input_ids": tf.int32, "attention_mask": tf.int32, "token_type_ids": tf.int32}, tf.int64),
        (
            {
                "input_ids": tf.TensorShape([None]),
                "attention_mask": tf.TensorShape([None]),
                "token_type_ids": tf.TensorShape([None]),
            },
            tf.TensorShape([]),
        ),
    )

# Définition des colonnes de données et d'étiquettes
DATA_COLUMN = 'review'
LABEL_COLUMN = 'sentiment'
```

#### Conversion des données en objets TF

```python
train_data = convertexamples2tf(list(train_examples), tokenizer)
train_data = train_data.shuffle(100).batch(32).repeat(2)
```

```python
validation_data = convertexamples2tf(list(validation_examples), tokenizer)
validation_data = validation_data.batch(32)
```

```python
test_data = convertexamples2tf(list(test_examples), tokenizer)
test_data = test_data.batch(32)
```

```python
model_Bert.compile(optimizer=tf.keras.optimizers.Adam(learning_rate=3e-5, epsilon=1e-08, clipnorm=1.0),
              loss=tf.keras.losses.SparseCategoricalCrossentropy(from_logits=True),
              metrics=[tf.keras.metrics.SparseCategoricalAccuracy('accuracy')])

model_Bert.fit(train_data, epochs=2, validation_data=validation_data)
```

```python
# Évaluation finale sur l'ensemble de test
# test_data = convertexamples2tf(test_examples, tokenizer)
evaluation = model_Bert.evaluate(test_data)
print("Perte (loss) sur les données de test :", evaluation[0])
print("Précision (accuracy) sur les données de test :", evaluation[1])
```

```python
import os,re
import seaborn as sns
from tqdm import tqdm
import pandas as pd
import numpy as np
import tensorflow as tf
print('tensorflow version:',tf.__version__)
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
from tensorflow.keras.layers import LSTM, InputLayer, Dense, Embedding, Input
from tensorflow.keras import Model
from sklearn.model_selection import train_test_split
import matplotlib.pylab as plt
```

```python
train_LSTM = pd.read_csv("/content/drive/MyDrive/MemoireM2/Notebook/Data_Imdb/train.csv")
test_LSTM = pd.read_csv("/content/drive/MyDrive/MemoireM2/Notebook/Data_Imdb/test.csv")
validation_LSTM = pd.read_csv("/content/drive/MyDrive/MemoireM2/Notebook/Data_Imdb/validation.csv")
```

```python
# Training set
x_train = np.array(train_LSTM['review'])
y_train = np.array(train_LSTM['sentiment'])

# Validation set
x_validation = np.array(validation_LSTM['review'])
y_validation = np.array(validation_LSTM['sentiment'])

# Test set
x_test = np.array(test_LSTM['review'])
y_test = np.array(test_LSTM['sentiment'])
```

```python
# Tokenization
tokenizer = Tokenizer(num_words=1000, filters='!"#$%&()*+,-./:;<=>?@[\\]^_`{|}~\t\n')
tokenizer.fit_on_texts(x_train)

x_train_matrix = tokenizer.texts_to_sequences(x_train)
x_validation_matrix = tokenizer.texts_to_sequences(x_validation)
x_test_matrix = tokenizer.texts_to_sequences(x_test)

```

```python
# Padding
maxlen = 32  # Vous pouvez choisir la longueur maximale des séquences selon vos données
x_train_pad = pad_sequences(x_train_matrix, maxlen=maxlen, padding='post')
x_validation_pad = pad_sequences(x_validation_matrix, maxlen=maxlen, padding='post')
x_test_pad = pad_sequences(x_test_matrix, maxlen=maxlen, padding='post')
```

```python
# Définition du modèle LSTM

# Taille du vocabulaire
vocab_length = len(tokenizer.word_index)

inputs = Input(shape=(maxlen,))
x = Embedding(input_dim=vocab_length + 1, output_dim=32, input_length=maxlen, embeddings_regularizer=tf.keras.regularizers.l2(0.001))(inputs)
x = LSTM(100, dropout=0.3, recurrent_dropout=0.2)(x)
outputs = Dense(1, activation='sigmoid')(x)
model = Model(inputs, outputs)


model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
# Entraînement du modèle
history = model.fit(x_train_pad, y_train, batch_size=64, epochs=10, verbose=1, shuffle=True, validation_data=(x_validation_pad, y_validation))
```

```python
# Évaluation du modèle sur l'ensemble de test
test_loss, test_accuracy = model.evaluate(x_test_pad, y_test, verbose=1)
print("Test Accuracy:", test_accuracy)
```

###LSTM With ATTENTION

```python
from tensorflow.keras.layers import Layer

class AttentionLayer(Layer):
    def __init__(self, **kwargs):
        super(AttentionLayer, self).__init__(**kwargs)

    def build(self, input_shape):
        self.W = self.add_weight(name="att_weight", shape=(input_shape[-1], 1), initializer="normal")
        self.b = self.add_weight(name="att_bias", shape=(input_shape[1], 1), initializer="zeros")
        super(AttentionLayer, self).build(input_shape)

    def call(self, x):
        et = tf.keras.backend.squeeze(tf.keras.backend.tanh(tf.keras.backend.dot(x, self.W) + self.b), axis=-1)
        at = tf.keras.backend.softmax(et, axis=1)
        at = tf.keras.backend.expand_dims(at, axis=-1)
        output = x * at
        return tf.keras.backend.sum(output, axis=1)

    def compute_output_shape(self, input_shape):
        return (input_shape[0], input_shape[-1])
```

```python
# Définition du modèle LSTM avec Attention
inputs = Input(shape=(maxlen,))
x = Embedding(input_dim=vocab_length + 1, output_dim=32, input_length=maxlen)(inputs)
x = LSTM(100, dropout=0.3, recurrent_dropout=0.2, return_sequences=True)(x)
x = AttentionLayer()(x)  # Couche d'attention
outputs = Dense(1, activation='sigmoid')(x)
model_LSTM_attention = Model(inputs, outputs)
```

```python
# Compilation
model_LSTM_attention.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
# Entraînement du modèle
history_LSTM_attention = model_LSTM_attention.fit(x_train_pad, y_train, batch_size=64, epochs=10, verbose=1, shuffle=True, validation_data=(x_validation_pad, y_validation))
```

```python
# Évaluation du modèle sur l'ensemble de test
test_loss_with_attention, test_accuracy_with_attention = model_LSTM_attention.evaluate(x_test_pad, y_test, verbose=1)
print("Test Accuracy (with attention):", test_accuracy_with_attention)
```
